## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
```

```python
plt.plot(x,y,color='green',linestyle='-.',marker='o')
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('learning')
```

```python
#subplots
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("plot1")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Plot 2")

plt.suptitle("subplots")
```
