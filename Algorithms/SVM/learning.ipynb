# Linear SVC 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000, n_features=2, n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
x
```

```python
pd.DataFrame(x).head()
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

# RBF kernel , Radial Basis Function ( Used when data is not linearly separable)

```python
#creating dataset
X,Y = make_classification(n_samples=1000, n_features=2, n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=Y)
```

```python
rbf = SVC(kernel='rbf')
```

```python
svc = SVC(kernel='linear')
```

```python
poly = SVC(kernel='poly')
```

```python
#train test split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
rbf.fit(x_train,y_train)
```

```python
svc.fit(x_train,y_train)
```

```python
poly.fit(x_train,y_train)
```

```python
y_pred = rbf.predict(x_test)
```

```python
y_pred2 = svc.predict(x_test)
```

```python
y_pred3 = poly.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred2))
```

```python
print(classification_report(y_test,y_pred3))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred2))
```

```python
print(confusion_matrix(y_test,y_pred3))
```

```python
train_accuracy = rbf.score(x_train, y_train)
test_accuracy = rbf.score(x_test, y_test)

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
#Hyper parameter tuning
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid =  {'C':[0.1,1,10,100,100],
               'gamma':[1,0.1,0.01,0.001,0.0001],
               'kernel':['rbf']}
grid = GridSearchCV(SVC(),param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(x_test)
print(classification_report(y_test,y_pred4))
```

```python
print(confusion_matrix(y_test,y_pred4))
```

```python
train_accuracy = grid.score(x_train, y_train)
test_accuracy = grid.score(x_test, y_test)

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```
