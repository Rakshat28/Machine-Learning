## Data manipulation and analysis with pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Value'] = df['Value'].fillna(df['Value'].mean())
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Sales']=df['Sales'].fillna(df['Sales'].mean())
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head()
```

```python
df=df.rename(columns={'Date':'Tareek'})
df.head()
```

```python
df=df.rename(columns={'Sales':'Bikri'})
df.head()
```

```python
df=df.rename(columns={'Value':'keemat'})
```

```python
df.head(3)
```

```python
df['Double Value']=df['keemat'].apply(lambda x:x*2)
df.head(3)
```

```python
df['New Bikri']=df['Bikri'].apply(lambda x:x*2)
df.head(3)
```

```python
grouped = df.groupby(['Category','Value'])['Bikri'].sum()
grouped
```

```python
df.head(4)
```

```python
grouped = df.groupby(['Product','Category'])['Sales'].agg(['sum','mean','max','min','size'])
grouped
```
